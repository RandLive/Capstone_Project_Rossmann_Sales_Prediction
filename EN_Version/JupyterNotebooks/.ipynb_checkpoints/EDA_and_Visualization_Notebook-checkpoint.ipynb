## Taking care of the outliers.

```python
df_train.head()
```

```python
# remove outliers and keep 0s
def remove_outliers(data):
    df_0 = data.loc[data.Sales ==0]   
    median = np.median(data.Sales)
    q1 = np.percentile(data.Sales, 25, axis=0)
    q3 = np.percentile(data.Sales, 75, axis=0)
    k = 1.5
    iqr = q3 - q1
    df_temp = data.loc[data.Sales > q1 - k*iqr]
    df_temp = data.loc[data.Sales < q3 + k*iqr]
#     frames = [df_0, df_temp]
#     result = pd.concat(frames)
    result = df_temp
    return result

df_train = remove_outliers(df_train)
```

## Handle the time-serie data

```python
import datetime as dt
time_format = '%Y-%m-%d'

def split_date(data):
    data_time = pd.to_datetime(data.Date, format=time_format)
    data['Year']= data_time.dt.year
    data['Month'] = data_time.dt.month
    data['DayOfYear'] = data_time.dt.dayofyear
    data['DayOfMonth'] = data_time.dt.day
    
    return data

df_train = split_date(df_train)
df_test = split_date(df_test)
```

```python
# Customer should be removed, because there is no such feature in the testing data
df_train = df_train[['Store', 'Sales', 'Year', 'Month', 'DayOfWeek', 'DayOfMonth',
                    'DayOfYear', 'Open', 'Promo', 'StateHoliday', 'SchoolHoliday','Date']]
df_test = df_test[['Store', 'Year', 'Month', 'DayOfWeek', 'DayOfMonth',
                    'DayOfYear', 'Open', 'Promo', 'StateHoliday', 'SchoolHoliday','Date']]
```

```python
# combine dataset
def combine_store(data):
    data = data.join(df_store, on='Store', rsuffix='_')
    data = data.drop('Store_',axis=1)
    return data

df_train = combine_store(df_train)
df_test = combine_store(df_test)
```

## str to int.

```python
# do some mappings in df
def do_map(data):
    
    data['StateHoliday'] = data['StateHoliday'].map({'0':0, 'a':1, 'b':2, 'c':3})
    data['StoreType'] = data['StoreType'].map({'0':0, 'a':1, 'b':2, 'c':3, 'd':4})
    data['Assortment'] = data['Assortment'].map({'0':0, 'a':1, 'b':2, 'c':3})
    data['PromoInterval'] = data['PromoInterval'].map({'0':0,'Jan,Apr,Jul,Oct':1,'Feb,May,Aug,Nov':2,'Mar,Jun,Sept,Dec':3})    
    
    return data

df_train = do_map(df_train)
df_test = do_map(df_test)
```

```python
# do some feature merging
def feature_merging(data):
    # create new features CompetitionLastMonth to describe 
    data['CompetitionLastMonths'] = 12 * (data['Year'] - data['CompetitionOpenSinceYear'].apply(lambda x: x if x > 0 else np.nan) - 1 \
                                          + data['CompetitionOpenSinceMonth'].apply(lambda x: x if x > 0 else np.nan))
   
    data['Promo2LastDays'] = 365 * (data['Year'] - data['Promo2SinceYear'].apply(lambda x: x if x > 0 else np.nan))/4.0 \
    + (data['DayOfYear'] - 7*(data['Promo2SinceWeek'].apply(lambda x: x if x > 0 else np.nan)) - 1)
    
    return data
```

```python
def feature_engineering_comb(data):
    # competition last 
    data['CompetitionLastMonths'] = 12 * (data['Year'] - data['CompetitionOpenSinceYear'].apply(lambda x: x if x > 0 else np.nan) - 1 \
                                          + data['CompetitionOpenSinceMonth'].apply(lambda x: x if x > 0 else np.nan))
   
    data['Promo2LastDays'] = 365 * (data['Year'] - data['Promo2SinceYear'].apply(lambda x: x if x > 0 else np.nan))/4.0 \
    + (data['DayOfYear'] - 7*(data['Promo2SinceWeek'].apply(lambda x: x if x > 0 else np.nan)) - 1)
    
    data = data.drop(['CompetitionOpenSinceMonth', 'CompetitionOpenSinceYear', 'Promo2SinceWeek', 'Promo2SinceYear'], axis=1)
    
    return data

df_train = feature_engineering_comb(df_train)
df_test = feature_engineering_comb(df_test)
```
